```python
import matplotlib.pyplot as plt
import numpy as np

def draw_line_variance(list1, list2, x_label, y_label, label1, label2, title, color1='b', color2='g'):
    sampling_steps = [0, 1, 2, 3, 4]
    list1_std = np.std(list1)
    list2_std = np.std(list2)

    # draw for Recall
    plt.figure(figsize=(9, 6))
    plt.plot(sampling_steps, list1, label=label1, color=color1, marker='o')
    plt.fill_between(sampling_steps, np.array(list1) - list1_std, 
                    np.array(list1) + list1_std, color=color1, alpha=0.2)

    plt.plot(sampling_steps, list2, label=label2, color=color2, marker='o')
    plt.fill_between(sampling_steps, np.array(list2) - list2_std, 
                    np.array(list2) + list2_std, color=color2, alpha=0.2)

    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.legend()
    plt.grid(True)
    # only show int step
    plt.xticks(np.arange(0, len(sampling_steps), 1))
    plt.show()
```

```python
x_label = 'Sampling Step'
y_label = 'Recall Value'
title = 'Comparison of Non-uniform and Uniform Sampling (Recall)'
label1 = 'Non-uniform Recall'
label2 = 'Uniform Recall'

tiktok_non_uniform_recall = [0.11452, 0.11256, 0.11223, 0.11354, 0.11615]
tiktok_uniform_recall = [0.11664, 0.11387, 0.11403, 0.11158, 0.10946]
# tiktok_non_uniform_recall = [0.11615, 0.11158, 0.11631, 0.11305, 0.11223]
# tiktok_uniform_recall = [0.11892, 0.11582, 0.11175, 0.11338, 0.11843]

draw_line_variance(
    tiktok_non_uniform_recall, tiktok_uniform_recall,
    x_label, y_label, label1, label2, title
)
```

```python
x_label = 'Sampling Step'
y_label = 'NDCG Value'
title = 'Comparison of Non-uniform and Uniform Sampling (NDCG)'
label1 = 'Non-uniform NDCG'
label2 = 'Uniform NDCG'

tiktok_non_uniform_ndcg = [0.04722, 0.04618, 0.04553, 0.04679, 0.04766]
tiktok_uniform_ndcg = [0.04872, 0.04785, 0.04733, 0.04625, 0.04563]
# tiktok_non_uniform_ndcg = [0.04900, 0.04597, 0.04712, 0.04703, 0.04645]
# tiktok_uniform_ndcg = [0.04940, 0.04819, 0.04639, 0.04667, 0.04863]

draw_line_variance(
    tiktok_non_uniform_ndcg, tiktok_uniform_ndcg, x_label, y_label, label1, label2, title,color1='r',color2='y'
)
```
